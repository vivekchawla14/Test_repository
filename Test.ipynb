```python
from pywinauto import Desktop

def window_exists(window_name: str) -> int:
    """
    Return 1 if a window containing `window_name` exists (case-insensitive), else 0.
    """
    target = window_name.strip().lower()
    desktop = Desktop(backend="win32")

    for w in desktop.windows():
        try:
            title = w.window_text()
            if title and target in title.strip().lower():
                return 1
        except Exception:
            pass

    return 0
```
